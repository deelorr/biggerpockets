```python
# four square method
# 1. Income = 
#     rental income = $2000.00
#     laundry 0
#     storage 0
#     misc 0
#     total monthly income = $2000.00
# 2. Expenses =
#     tax = $150.00
#     insurance = $100.00
#     utilities = 0
#         electric:
#         water:
#         sewer:
#         garbage:
#         gas:
#     HOA = 0
#     lawn/snow = 0
#     vacancy = $100
#     repairs = $100  
#     capEx = $100
#     property managemnet = $200
#     mortgage = $860
#     total monthly expenses = $1610
# 3. Cash Flow = 
#     income - expenses = cash Flow
#     2000-1610=390
#     total monthly cashflow = 390
# 4. cash on cash ROI = 
#     down payment = $40000
#     closing costs = $3000
#     rehab budget = $7000
#     misc other = 
#     total investment = $50000
#     $390*12 = $4680 annual cash Flow
#     annual cash flow / total investment
#     9.36% cash on cash ROI

class CalcROI:
    def __init__(self):
        pass
    
    def income(self, rental_income, laundry=0, storage=0, misc=0):
        total = rental_income + laundry + storage + misc
        return total
    
    def expenses(self, tax, ins, util=0, hoa=0, lawn=0, vac=0, rep=0, capEx=0, pman=0, mort=0):
        total = tax + ins + util + hoa + lawn + vac + rep + capEx + pman + mort
        return total
    
    def cashflow(self, total_income, total_expenses):
        cash_flow = total_income - total_expenses
        annual_cashflow = cash_flow * 12
        return annual_cashflow

    def roi(self, downpay, closing, rehab, misc=0):
        total_investment = downpay + closing + rehab + misc
        annual_cashflow = self.cashflow(total_income, total_expenses)
        roi = round(annual_cashflow / total_investment * 100, 2)
        return roi

calc = CalcROI()
needed_values = []

total_income = calc.income(2000)
needed_values.append(total_income)

total_expenses = calc.expenses(150, 100, vac=100, rep=100, capEx=100, pman=200, mort=860)
needed_values.append(total_expenses)

cash_flow = calc.cashflow(total_income, total_expenses)
needed_values.append(cash_flow)

cashroi = calc.roi(40000, 300, 7000)
needed_values.append(cashroi)

# print(needed_values)
print(f"Monthly Income: {needed_values[0]}")
print(f"Monthly Expenses: {needed_values[1]}")
print(f"Yearly Cashflow: {needed_values[2]}")
print(f"Cash on Cash ROI: {needed_values[3]}%")
```
